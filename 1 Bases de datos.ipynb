Contruir un par de tablas en la BD wORK de nombre E8SQL y E8SAS que contengan los siguientes campos: Cliente, Fecha de Nacimiento, Ingreso Total, Frecuencia de Pago, Sexo, Estado.

**Ejercicios c4**

Crear un par de tablas de nombre E9SQL y E9SAS que contengan los siguientes campos:
Cliente, Nombre, Apellido Paterno, Fecha de Nacimiento, Fecha en que se hizo la solicitud,
Edad, Ingreso Total. Aplicando los siguienes filtros: Ingreso Total > 10000 y Edad > 24

```python
temp = pd.merge(nombre,demograficos,on="Cliente",how="left").copy()
temp = pd.merge(temp,solicitud,on="Cliente",how="left").copy()
temp["Edad"] = ((pd.Timestamp.now() - temp["Fecha_Nacimiento"]).dt.days / 365.25).astype(int)
temp["Ingreso_Total"] = temp["Ingreso_Fijo"] + temp["Ingreso_Variable"]
filtro = (temp["Ingreso_Total"] > 10000) & (temp["Edad"] > 24)
temp[filtro][["Cliente","Nombre_Solic","ApPaterno_Solic","Fecha_Nacimiento","Fecha_Solicitud","Edad","Ingreso_Total"]]
```

MINI EXAMEN

Crear un par de tablas de nombre E10SQL y E10SAS que contengan los siguientes campos:
Cliente, Nombre, Apellido Paterno, Monto Solicitado, Ingreso Total, Edad, Estatus y BC Score.
Aplicar los siguientes filtros: Ingreso Total > 10000, Mas de 40 años de edad, Vivir en el DF, Solo clientes
aprobados (ie estatus = "A") y BC Score > 650

```python
temp = pd.merge(nombre,demograficos,on="Cliente",how="left").copy()
temp = pd.merge(temp,solicitud,on="Cliente",how="left").copy()
temp = pd.merge(temp,buro,on="Cliente",how="left").copy()
temp = pd.merge(temp,aprobadas,on="Cliente",how="left").copy()
temp["Edad"] = ((pd.Timestamp.now() - temp["Fecha_Nacimiento"]).dt.days / 365.25).astype(int)
temp["Ingreso_Total"] = temp["Ingreso_Fijo"] + temp["Ingreso_Variable"]
filtro = (temp["Ingreso_Total"] > 10000) & (temp["Edad"] > 40) & (temp["Estado"]=="DISTRITO FEDERAL") & (temp["ESTATUS"]=="A") & (temp["BC_Score"] > 650)
temp[filtro][["Cliente","Nombre_Solic","ApPaterno_Solic","Monto_Solicitado","Edad","Ingreso_Total","Edad","ESTATUS","BC_Score"]]
```
